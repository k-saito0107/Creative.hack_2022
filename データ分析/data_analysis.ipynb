```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta
import random
import glob

# スクレイピング結果の保存先
file_dir_path = '../../data/'
# 乱数seed固定
random.seed(1234) 
```

```python
# スクレイピング結果の読み込み
csv_file_list = glob.glob(file_dir_path + '*.csv')
sample_csv_file = pd.read_csv(csv_file_list[0])
print(len(sample_csv_file))
sample_csv_file.head()
```

```python
# 取得したスクレイピング結果を一つのcsvにまとめる
# 公式アカウントによるツイートの削除

for i in range(len(csv_file_list)):
    read_data = pd.read_csv(csv_file_list[i])
    read_data = read_data[~read_data['自己紹介文'].str.contains('公式',na=False)]
    #read_data = read_data[~read_data['ツイート本文'].str.contains('RT',na=False)]
    if i == 0:
        data = read_data
    else:
        data = pd.concat([data, read_data], axis=0, ignore_index=True)
data = data.drop_duplicates(subset='ツイートID')
data = data.fillna('None')
```

```python
data.head()
```

```python
# データの整形

# ツイートの文字数を計測
def count_string_length(df):
    result = len(df['ツイート本文'])
    return result

# ヘッダー画像の有無
def check_header_image(df):
    if 'http' in df['ヘッダー画像URL']:
        result = 1
    else:
        result = 0
    return result

# ヘッダーに埋め込まれたリンクの有無
def check_header_url(df):
    if 'http' in df['WEBサイト']:
        result = 1
    else:
        result = 0
    return result

# 日付整形
def date_modify(df):
    date = df.split('_')[0]
    date = pd.to_datetime(date)
    return date

data['文字数'] = data.apply(count_string_length,axis=1)
data['ヘッダー画像の有無'] = data.apply(check_header_image,axis=1)
data['ヘッダーURLの有無'] = data.apply(check_header_url,axis=1)
data['ツイート時刻'] = data['ツイート時刻'].apply(date_modify)
data['アカウント作成日時'] = data['アカウント作成日時'].apply(date_modify)
data['アカウント稼働日数'] = data['ツイート時刻'] - data['アカウント作成日時']
data['アカウント稼働日数'] = (data['アカウント稼働日数'] / timedelta(days=1))
```

```python
data = data.drop('アイコン画像URL', axis=1)
data = data.drop('ヘッダー画像URL', axis=1)
data = data.drop('WEBサイト', axis=1)
data = data.drop('自分のフォロー状況', axis=1)
```

```python
save_path = 'reshape_data.csv'
data.to_csv(save_path)
```
