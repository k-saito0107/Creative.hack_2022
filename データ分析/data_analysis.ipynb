```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta
import random
import glob

# ã‚¹ã‚¯ãƒ¬ã‚¤ãƒ”ãƒ³ã‚°çµæœã®ä¿å­˜å…ˆ
file_dir_path = '../../data/'
# ä¹±æ•°seedå›ºå®š
random.seed(1234) 
```

```python
# ã‚¹ã‚¯ãƒ¬ã‚¤ãƒ”ãƒ³ã‚°çµæœã®èª­ã¿è¾¼ã¿
csv_file_list = glob.glob(file_dir_path + '*.csv')
sample_csv_file = pd.read_csv(csv_file_list[0])
print(len(sample_csv_file))
sample_csv_file.head()
```

```python
# å–å¾—ã—ãŸã‚¹ã‚¯ãƒ¬ã‚¤ãƒ”ãƒ³ã‚°çµæœã‚’ä¸€ã¤ã®csvã«ã¾ã¨ã‚ã‚‹
# å…¬å¼ã‚¢ã‚«ã‚¦ãƒ³ãƒˆã«ã‚ˆã‚‹ãƒ„ã‚¤ãƒ¼ãƒˆã®å‰Šé™¤

for i in range(len(csv_file_list)):
    read_data = pd.read_csv(csv_file_list[i])
    read_data = read_data[~read_data['è‡ªå·±ç´¹ä»‹æ–‡'].str.contains('å…¬å¼',na=False)]
    #read_data = read_data[~read_data['ãƒ„ã‚¤ãƒ¼ãƒˆæœ¬æ–‡'].str.contains('RT',na=False)]
    if i == 0:
        data = read_data
    else:
        data = pd.concat([data, read_data], axis=0, ignore_index=True)
data = data.drop_duplicates(subset='ãƒ„ã‚¤ãƒ¼ãƒˆID')
data = data.fillna('None')
```

```python
# ãƒ‡ãƒ¼ã‚¿ã®æ•´å½¢

# ãƒ„ã‚¤ãƒ¼ãƒˆã®æ–‡å­—æ•°ã‚’è¨ˆæ¸¬
def count_string_length(df):
    result = len(df['ãƒ„ã‚¤ãƒ¼ãƒˆæœ¬æ–‡'])
    return result

# ãƒ˜ãƒƒãƒ€ãƒ¼ç”»åƒã®æœ‰ç„¡
def check_header_image(df):
    if 'http' in df['ãƒ˜ãƒƒãƒ€ãƒ¼ç”»åƒURL']:
        result = 1
    else:
        result = 0
    return result

# ãƒ˜ãƒƒãƒ€ãƒ¼ã«åŸ‹ã‚è¾¼ã¾ã‚ŒãŸãƒªãƒ³ã‚¯ã®æœ‰ç„¡
def check_header_url(df):
    if 'http' in df['WEBã‚µã‚¤ãƒˆ']:
        result = 1
    else:
        result = 0
    return result

# æ—¥ä»˜æ•´å½¢
def date_modify(df):
    date = df.split('_')[0]
    date = pd.to_datetime(date)
    return date

data['æ–‡å­—æ•°'] = data.apply(count_string_length,axis=1)
data['ãƒ˜ãƒƒãƒ€ãƒ¼ç”»åƒã®æœ‰ç„¡'] = data.apply(check_header_image,axis=1)
data['ãƒ˜ãƒƒãƒ€ãƒ¼URLã®æœ‰ç„¡'] = data.apply(check_header_url,axis=1)
data['ãƒ„ã‚¤ãƒ¼ãƒˆæ™‚åˆ»'] = data['ãƒ„ã‚¤ãƒ¼ãƒˆæ™‚åˆ»'].apply(date_modify)
data['ã‚¢ã‚«ã‚¦ãƒ³ãƒˆä½œæˆæ—¥æ™‚'] = data['ã‚¢ã‚«ã‚¦ãƒ³ãƒˆä½œæˆæ—¥æ™‚'].apply(date_modify)
data['ã‚¢ã‚«ã‚¦ãƒ³ãƒˆç¨¼åƒæ—¥æ•°'] = data['ãƒ„ã‚¤ãƒ¼ãƒˆæ™‚åˆ»'] - data['ã‚¢ã‚«ã‚¦ãƒ³ãƒˆä½œæˆæ—¥æ™‚']
data['ã‚¢ã‚«ã‚¦ãƒ³ãƒˆç¨¼åƒæ—¥æ•°'] = (data['ã‚¢ã‚«ã‚¦ãƒ³ãƒˆç¨¼åƒæ—¥æ•°'] / timedelta(days=1))
```

```python
data = data.drop('ã‚¢ã‚¤ã‚³ãƒ³ç”»åƒURL', axis=1)
data = data.drop('ãƒ˜ãƒƒãƒ€ãƒ¼ç”»åƒURL', axis=1)
data = data.drop('WEBã‚µã‚¤ãƒˆ', axis=1)
data = data.drop('è‡ªåˆ†ã®ãƒ•ã‚©ãƒ­ãƒ¼çŠ¶æ³', axis=1)
```
